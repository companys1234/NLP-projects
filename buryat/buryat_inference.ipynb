```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel

# Просто укажите ваш ID репозитория с Hugging Face
hf_model_id = "ваше_имя_пользователя/llama-3.2-1b-ru-bua-translator"
base_model_id = "unsloth/Llama-3.2-1B" # Оригинальный чистый фундамент

# 1. Скачиваем токенизатор прямо с вашего хаба
tokenizer = AutoTokenizer.from_pretrained(hf_model_id)

# 2. Скачиваем чистую базу
base_model = AutoModelForCausalLM.from_pretrained(
    base_model_id, torch_dtype=torch.float16, device_map="auto"
)

# 3. 🔥 СКАЧИВАЕМ И НАКАТЫВАЕМ ВАШИ БУРЯТСКИЕ ВЕСА ИЗ ИНТЕРНЕТА 🔥
model = PeftModel.from_pretrained(base_model, hf_model_id)
model.eval()

# Всё готово к генерации!
```
